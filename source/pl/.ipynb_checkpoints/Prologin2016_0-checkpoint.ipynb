## 1

```
On note « A en base b » : A_b. Combien font, en base 10, 10_2 + 10_8 + 10_16 ?

Notez uniquement la valeur en base 10, sans préciser la base.
```

10 en base 2, c'est 2 en base 10
10 en base 8, c'est 8 en base 10
10 en base 16, c'est 16 en base 10

donc, la solution est 2 + 8 + 16 = 26

Quel est le 1000ème entier qui a un nombre premier de 1 dans sa représentation binaire ? Rappel : 1 n'est pas premier.
```


```
nombre en base 10 =   n_2   => nb chiffre  1 = entier / ou pas  =  ordre
    1  =  2^1 - 1 =  00001  =>  1 chiffre  1 = Non (1 pas prem)
    2  =          =  00010  =>  1 chiffre  1 = Non (1 pas prem)
    3  =  2^2 - 1 =  00011  =>  2 chiffres 1 = Oui              =  1
    4  =          =  00100  =>  1 chiffre  1 = Non (1 pas prem)
    5  =          =  00101  =>  2 chiffres 1 = Oui              =  2
    6  =          =  00110  =>  2 chiffres 1 = Oui              =  3
    7  =  2^3 - 1 =  00111  =>  3 chiffres 1 = Oui              =  4
    8  =          =  01000  =>  1 chiffre  1 = Non (1 pas prem)
    9  =          =  01001  =>  2 chiffres 1 = Oui              =  5
   10  =          =  01010  =>  2 chiffres 1 = Oui              =  6
   11  =          =  01011  =>  3 chiffres 1 = Oui              =  7
   12  =          =  01100  =>  2 chiffres 1 = Oui              =  8
   13  =          =  01101  =>  3 chiffres 1 = Oui              =  9
   14  =          =  01110  =>  3 chiffres 1 = Oui              = 10
   15  =  2^4 - 1 =  01111  =>  4 chiffres 1 = Non (4 pas prem)
   16  =          =  10000  =>  1 chiffre  1 = Non (1 pas prem)
   17  =          =  10001  =>  2 chiffres 1 = Oui              = 11
   18  =          =  10010  =>  2 chiffres 1 = Oui              = 12
   19  =          =  10011  =>  3 chiffres 1 = Oui              = 13
   20  =          =  10100  =>  2 chiffres 1 = Oui              = 14
   21  =          =  10101  =>  3 chiffres 1 = Oui              = 15
   22  =          =  10110  =>  3 chiffres 1 = Oui              = 16
   23  =          =  10111  =>  4 chiffres 1 = Non (4 pas prem)
   24  =          =  11000  =>  2 chiffres 1 = Oui              = 17
   25  =          =  11001  =>  3 chiffres 1 = Oui              = 18
   26  =          =  11010  =>  3 chiffres 1 = Oui              = 19
   27  =          =  11011  =>  4 chiffres 1 = Non (4 pas prem)
   28  =          =  11100  =>  3 chiffres 1 = Oui              = 20
   29  =          =  11101  =>  4 chiffres 1 = Non (4 pas prem)
   30  =          =  11110  =>  4 chiffres 1 = Non (4 pas prem)
   31  =  2^5 - 1 =  11111  =>  5 chiffres 1 = Oui              = 21
```
Creation de ce tableau en python :
 - 1ere colonne : nombre en décimal
 - 2ème colonne : ce meme nombre en écriture binaire
 - 3ème colonne : nombre de chiffre 1 dans l'écriture binaire
 - 4ème colonne : est ce que ce nombre de chiffre 1 est un nombre entier
 - 5ème colonne : compteur pour s'arreter au 1000ème nombre qui correpond au critère


Le fichier est :
```
ex3_1000premiers.py
```

Le résultat est :
```
1990
```




## 4
```
